```python
from PIL import Image
import PIL.Image
import cv2 
import matplotlib.pyplot as plt
import datetime
import numpy as np
from PIL import Image, ImageEnhance 
from pytesseract import image_to_string
import pytesseract
import pandas as pd

pytesseract.pytesseract.tesseract_cmd = 'Tesseract-OCR/tesseract'
TESSDATA_PREFIX = 'Tesseract-OCR'
```

```python
import os
arr = os.listdir('DataProj/')
print(arr)
```

```python
labels = ['Name','Chasis No.','Engine No.','Registration No.','Registration Date','Model']
data = np.zeros((47,6),dtype = str)
data = pd.DataFrame(data)
data.columns = labels
```

```python
q = 0
name = []
chasis = []
symbols = [',','.','/','?','*','<','>',':','[','{',']','}','!','@','#','$','%','^','&','(',')','-','_','=','+','-:','>:','=:','_.',';','__:','—_:','_:','=<:']
l = []
for i in arr[1:]:
    im = Image.open("DataProj/"+i)
    enhancer = ImageEnhance.Brightness(im)
    enhanced_im = enhancer.enhance(1.5)
    output = pytesseract.image_to_string(enhanced_im, lang='eng')
    outlist = output.strip().split()
    for j in outlist:
        if j not in symbols:
            l.append(j.strip())
    name = ""
    w = ""
    e = ""
    r = ""
    for k in range(len(l)):
        
        if l[k]=='NAME' or l[k]=='Name.':
            if l[k+1] == "‘S/WID" or l[k+1] == "SMD" or l[k+1] == "‘SIMWID" or l[k+1] == "SIMWID" or l[k+1] == "SMWID" or l[k+1]=="Vehicle" or l[k+1]=="Hip" or l[k+1]=="SIWID" or l[k+1]=="S/D/W/o":
                break
            else:
                
                w = l[k+1]
                
                
                
                if l[k+2].startswith('SIW') or l[k+2].startswith('AD'):
                    break
                else:
                    
            
                    e = l[k+2]
                   
                    if l[k+3].startswith('‘SI') or l[k+3].startswith('AD') or  l[k+3].startswith('S') or l[k+3].startswith('ru') :
                        break
                    else:
                       
                        r = l[k+3]
                        
        if len(l[k]) == 17:
            
            data['Chasis No.'].iloc[q] = l[k]
        if l[k] == 'ENO':
            data['Engine No.'].iloc[q] = l[k+1]
            
        if l[k] == 'Engine' and (l[k+1] != "No." or l[k+1] != "No.."):
            data['Engine No.'].iloc[q] = l[k+1]
        if l[k] == 'Engine' and (l[k+1] == "No." or l[k+1] == "No.."):
            data['Engine No.'].iloc[q] = l[k+2]
        if l[k] == "Registration" and l[k+2] != "No.":
            data['Registration No.'].iloc[q] = l[k+2]
        if l[k] == 'REGN':
            data['Registration No.'].iloc[q] = l[k+2]
        if l[k] == 'REG.' or l[k] == 'Manufacture':
            data['Registration Date'].iloc[q] = l[k+2]
        if l[k] == 'MODEL':
            data['Model'].iloc[q] = l[k+1]+" "+l[k+2]
        if l[k] == 'of' and l[k+1] == 'Vehicle':
            data['Model'].iloc[q] = l[k+2]
            

    name = w + " "+ e + " " + r
    data['Name'].iloc[q] = name
    q+=1
    l=[]
```

```python
data.to_csv("data.csv")
```
